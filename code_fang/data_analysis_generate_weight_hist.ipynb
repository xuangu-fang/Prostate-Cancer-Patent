```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
# sns.set_style("darkgrid")
```

```python
def name_replace(string_list):
    return [string.replace("SNP", "Mut") for string in string_list]
    


# figs of weight analysis:P1

target_P1 = ['1= death; 0=alive','ADT_if_fail']
target_P2 = ['gap_surv_time_class'+'_early_death','gap_surv_time_class'+'_long_survive']

P1_effect_in = []
P1_effect_de = []

P1_feature_in = []
P1_feature_de = []

P2_effect_in = []
P2_effect_de = []

P2_feature_in = []
P2_feature_de = []

P1_top = 20
P2_top = 20


for target in target_P1:


    file_name_de = './result_log/P1_weight_'+target+'_de_'+'.csv'
    file_name_in = './result_log/P1_weight_'+target+'_in_'+'.csv'

    table_de = pd.read_csv(file_name_de)
    table_in = pd.read_csv(file_name_in)


    feature_de = table_de['feature_name'][:P1_top].values.tolist()
    feature_de = name_replace(feature_de)
    # effct_de = table_de['relative_effects_ratio'][:15]
    effct_de = table_de['feature_weight'][:P1_top]/table_de['feature_weight'].sum()

    feature_in = table_in['feature_name'][:P1_top].values.tolist()
    feature_in = name_replace(feature_in)
    # effct_in = table_in['relative_effects_ratio'][:15]
    effct_in = table_in['feature_weight'][:P1_top]/table_in['feature_weight'].sum()

    P1_effect_in.append(effct_in)
    P1_effect_de.append(effct_de)

    P1_feature_in.append(feature_in)
    P1_feature_de.append(feature_de)


for target in target_P2:
    

    file_name_de = './result_log/P2_weight_'+target+'_de_binary'+'.csv'
    file_name_in = './result_log/P2_weight_'+target+'_in_binary'+'.csv'

    table_de = pd.read_csv(file_name_de)
    table_in = pd.read_csv(file_name_in)


    feature_de = table_de['feature_name'][:P2_top].values.tolist()
    feature_de = name_replace(feature_de)
    # effct_de = table_de['relative_effects_ratio'][:15]
    effct_de = table_de['feature_weight'][:P2_top]/table_de['feature_weight'].sum()

    feature_in = table_in['feature_name'][:P2_top].values.tolist()
    feature_in = name_replace(feature_in)
    # effct_in = table_in['relative_effects_ratio'][:15]
    effct_in = table_in['feature_weight'][:P2_top]/table_in['feature_weight'].sum()

    P2_effect_in.append(effct_in)
    P2_effect_de.append(effct_de)

    P2_feature_in.append(feature_in)
    P2_feature_de.append(feature_de)
```

```python
f = plt.figure(figsize=(16,8))

y_lim = 0.13

ax1 = f.add_subplot(141)
plt.ylabel('Relative Effct Ratio', axes=ax1,fontsize=22)
plt.bar(range(P1_top),P1_effect_de[0],axes=ax1,tick_label=P1_feature_de[0],color=(1,0.3,0))
plt.yticks(np.arange(0, y_lim, step=0.02),fontsize=10,axes=ax1)
plt.xticks(rotation =90,axes=ax1,fontsize=10)
plt.title('mHSPC:Survival',fontsize=20,axes=ax1)

ax2 = f.add_subplot(142)
plt.bar(range(P1_top),P1_effect_de[1],axes=ax2,tick_label=P1_feature_de[1],color=(0,0,0.55))
plt.yticks(np.arange(0, y_lim, step=0.02),fontsize=10,axes=ax2)
plt.xticks(rotation =90,axes=ax2)
plt.title('mHSPC:ADT Failure',fontsize=20,axes=ax2)

ax3 = f.add_subplot(143)
plt.bar(range(P2_top),P2_effect_de[0],axes=ax3,tick_label=P2_feature_de[0],color=(0/255,139/255,139/255))
plt.yticks(np.arange(0,y_lim, step=0.02),fontsize=10,axes=ax3)
plt.xticks(rotation =90,axes=ax3)
plt.title('mCRPC:Early Death',fontsize=20,axes=ax3)

ax4 = f.add_subplot(144)
plt.bar(range(P2_top),P2_effect_de[1],axes=ax4,tick_label=P2_feature_de[1],color=(138/255,43/255,226/255))
plt.yticks(np.arange(0, y_lim, step=0.02),fontsize=10,axes=ax4)
plt.title('mCRPC:Long Survival',fontsize=20,axes=ax4)

plt.xticks(rotation =90,axes=ax4)


```
