```python
# analysis the t-test and p-value
from scipy.stats import ttest_ind
v1 = np.random.normal(size=10)
v2 = np.random.normal(size=100)

res = ttest_ind(v1, v2).pvalue
res
```

```python
test_table1 = p1_spicie_cont_single
test_table2 = p2_spicie_cont_single

test_var1 = P1_Cer_names
test_var2 = P2_Cer_names

var_name_list = []
p_value_list = []

for i in range(len(test_var1)):
    var_name =  test_var1[i].split('_')[-1]
    var_name_list.append(var_name)

    var1 = test_table1[test_var1[i]]
    var2 = test_table2[test_var2[i]]

    p_value_list.append(ttest_ind(var1, var2).pvalue)
```

```python
Cer_lip_diff_table = pd.DataFrame()
Cer_lip_diff_table['cer_lip_name'] = np.array(var_name_list)
Cer_lip_diff_table['p_value'] = np.array(p_value_list)
Cer_lip_diff_table_sort = Cer_lip_diff_table.sort_values('p_value')
Cer_lip_diff_table_sort.to_csv('../data/processed/Cer_lip_diff_table_sort.csv',index=False)
```
