```python
group_name_set = set(group_name_list)
```

```python
group1 = ['BAP1','FBXW7','AKT1','TP53','TSC2','CDKN2A']
group2 = ['BRCA1','RB1','AKT','CDK4','PRKD1','PDGFRA','IDH1','FGFR3','MTOR','KIT'] # duplicate 'FBXW7' in two groups
```

```python
table_gene_new = pd.DataFrame()
table_gene_new['Patient_ID'] = gene_table['Patient_ID']

for gene_name in gene_names_list:
    df = gene_table[gene_name]

    gene_group_name = gene_name.split('_')[0]
    group_name = '2'
    if gene_group_name in group1:
        group_name = '0'
    elif gene_group_name in group2:
        group_name = '1'
    table_gene_new['gene_' + gene_name + '_' + group_name] = df.fillna(0)
table_gene_new  
```
