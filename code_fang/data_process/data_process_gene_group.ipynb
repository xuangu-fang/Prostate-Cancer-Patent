```python
gene_table = pd.read_excel('../data/raw/processed_gene_full.xlsx')
# lip_name_table = pd.read_excel('../data/raw/LCMS_data_copy.xlsx',4)
```

```python
gene_names_list = list(gene_table.columns) # id and features name
gene_names_list = gene_names_list[1:-2]
# gene_names_list
```

```python
gene_names_list
```

```python
len(gene_names_list)
```

```python
group_name_list = [item.split('_')[0] for item in gene_names_list]# get group name for each specie
```

```python
class_name_set = set(group_name_list)
len(set(class_name_set))
```

```python
class_list = list(class_name_set)
textfile = open("class_list_file_gene.txt", "a+")
for element in class_list:
    textfile.write(element + "\n")
textfile.close()
```

```python
# table/dic2: class_name->group_name
class_group_dict={}
for bio_class in class_list:
    matched_groups_list = [gene_names_list[i] for i in range(len(gene_names_list)) if group_name_list[i]==bio_class]
    class_group_dict[bio_class] = matched_groups_list
```

```python
table_gene_group = pd.DataFrame()
table_gene_group['Patient_ID'] = gene_table['Patient_ID']

for bio_class_name in class_list:
    df = gene_table[class_group_dict[bio_class_name]]

    table_gene_group['gene_' + bio_class_name] = (df.sum(axis=1)>0)*1 # if there any sub-group>1(has mutation), assign 1, else assign 0
table_gene_group
```
