```python
table_gene_group = pd.DataFrame()
table_gene_group['Patient_ID'] = gene_table['Patient_ID']

for bio_class_name in class_list:
    df = gene_table[class_group_dict[bio_class_name]]

    table_gene_group['gene_' + bio_class_name] = (df.sum(axis=1)>0)*1 # if there any sub-group>1(has mutation), assign 1, else assign 0
table_gene_group
# table_gene_group.to_csv('../data/processed/process_gene_group.csv',index=False)
```

```python
# pathway check: to be checked in the future
pathway_list = []

# textfile = open("../data/pathway/can_top100.txt", "r")
textfile = open("../data/pathway/diseases_top100.txt", "r")


for line in textfile:
    pathway_raw = line.strip().split(',')
    pathway = ['gene_'+gene for gene in pathway_raw if gene in class_list ]  # filter out the gene classes that don't show in our data
    pathway_list.append(pathway)

pathway_list
```

```python
# (gene_table[pathway_list[0]].sum(axis=1)>0)*1
act_ratio = []
for i in range(len(pathway_list)):
    n = len(pathway_list[i])
    max_v = (table_gene_group[pathway_list[i]].sum(axis=1)).max()
    med_v = (table_gene_group[pathway_list[i]].sum(axis=1)).quantile(q=0.76)
    act_ratio.append(med_v/n)
    print(n,max_v,med_v)
np.array(act_ratio).mean()
```

```python
table_gene_pathway = pd.DataFrame()
table_gene_pathway['Patient_ID'] = gene_table['Patient_ID']

for i in range(len(pathway_list)):

    n = len(pathway_list[i])
    df = table_gene_group[pathway_list[i]]
    med_v = (table_gene_group[pathway_list[i]].sum(axis=1)).quantile(q=0.66)
    thr = med_v

    table_gene_pathway['pw_' + str(i)] = (df.sum(axis=1)>thr)*1 # 
table_gene_pathway
```
