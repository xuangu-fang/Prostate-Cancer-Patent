```python
# table/dic1: group_name->class_name
group_class_dict={}
for i in range(len(feature_list_specie)):
    group_class_dict[feature_list_specie[i]] = group_name_list[i]
group_class_dict 
```

```python
# table/dic2: class_name->group_name
class_group_dict={}
for bio_class in class_list:
    matched_groups_list = [feature_list_specie[i] for i in range(len(group_name_list)) if group_name_list[i]==bio_class]
    class_group_dict[bio_class] = matched_groups_list
```

```python
def id_process(id):

    if id[-1] == 'C':
        # pattern 'num + C' 
        return id[:-2]
    elif id[1] == '-':
        # pattern 'P-num' 
        return id[2:]
    else:# pattern 'P num'
        return id[1:]
table_specie.iloc[:,2].apply(id_process)
```

```python
# V1: group-wise lip feature: first standarize, then sum, then binarize
#  if val > median, set feature_high = 1,feature_low = 0
#  else,   set feature_high = 0,feature_high = 1

table_specie_group_bi = pd.DataFrame()
table_specie_group_bi['Patient_ID'] = table_specie.iloc[:,2].apply(id_process)

for bio_class_name in class_list:
    df = table_specie[class_group_dict[bio_class_name]]
    normalized_df = (df-df.min())/(df.max()-df.min())
    sum_df = normalized_df.sum(axis=1)
    # normalized_df = (df-df.mean())/df.std()
    # normalized_sum_df = (sum_df-sum_df.min())/(sum_df.max()-sum_df.min())

    # group_name = name_group_link.loc[name_group_link['Updated name']==lip_name]['Class'].values[0]
    # group_name_idx = str(group_name_dict[group_name])
    table_specie_group_bi['lip_' + bio_class_name +'_'+'high'] =  (sum_df>=sum_df.median())*1
    table_specie_group_bi['lip_' + bio_class_name +'_'+'low'] = (sum_df<sum_df.median())*1 

table_specie_group_bi   
# table_specie_group_bi.to_csv('../data/processed/process_lip_group_binary_v1.csv',index=False)
```

```python
# V2: group-wise lip feature: first sum, then standarize. than binarize (no first standarize for each sub-group lip compare to V1)
#  if val > median, set feature_high = 1,feature_low = 0
#  else,   set feature_high = 0,feature_high = 1

table_specie_group_bi = pd.DataFrame()
table_specie_group_bi['Patient_ID'] = table_specie.iloc[:,2].apply(id_process)

for bio_class_name in class_list:
    df = table_specie[class_group_dict[bio_class_name]]
    # normalized_df = (df-df.min())/(df.max()-df.min())
    sum_df = df.sum(axis=1)
    # normalized_df = (df-df.mean())/df.std()
    # normalized_sum_df = (sum_df-sum_df.min())/(sum_df.max()-sum_df.min())

    # group_name = name_group_link.loc[name_group_link['Updated name']==lip_name]['Class'].values[0]
    # group_name_idx = str(group_name_dict[group_name])
    table_specie_group_bi['lip_' + bio_class_name +'_'+'high'] =  (sum_df>=sum_df.median())*1
    table_specie_group_bi['lip_' + bio_class_name +'_'+'low'] = (sum_df<sum_df.median())*1 

table_specie_group_bi   
# table_specie_group_bi.to_csv('../data/processed/process_lip_group_binary_v2.csv',index=False)
```

```python
# V3: group-wise lip feature: first standarize, then sum, then standarize

table_specie_group = pd.DataFrame()
table_specie_group['Patient_ID'] = table_specie.iloc[:,2].apply(id_process)

for bio_class_name in class_list:
    df = table_specie[class_group_dict[bio_class_name]]
    normalized_df = (df-df.min())/(df.max()-df.min())
    sum_df = normalized_df.sum(axis=1)
    # normalized_df = (df-df.mean())/df.std()
    normalized_sum_df = (sum_df-sum_df.min())/(sum_df.max()-sum_df.min())

    # group_name = name_group_link.loc[name_group_link['Updated name']==lip_name]['Class'].values[0]
    # group_name_idx = str(group_name_dict[group_name])
    table_specie_group['lip_' + bio_class_name ] =  normalized_sum_df
    

table_specie_group
table_specie_group.to_csv('../data/processed/process_lip_group_v3.csv',index=False)
```
